#### Emily Barraza NL = 7
#### Problema Mecánico
* Taller de trabajo
* especialidad de motores
* herramientas principales
* certificaciones
* reparaciones diarias
* años de oficio.

```python
# =========================================
# PROGRAMA: TALLER MECÁNICO INTELIGENTE
# Ejemplo de Programación Orientada a Objetos
# =========================================

class Mecanico():

    # Constructor
    def __init__(self, taller, especialidad, herramientas,
                 certificaciones, reparaciones, anios):

        self.taller = taller
        self.especialidad = especialidad
        self.herramientas = herramientas
        self.certificaciones = certificaciones
        self.reparaciones = reparaciones
        self.anios = anios

    # Mostrar información del mecánico
    def mostrar_datos(self):

        print(f"{self.taller:<20} | "
              f"{self.especialidad:<20} | "
              f"{self.herramientas:<20} | "
              f"{self.certificaciones:<20} | "
              f"{self.reparaciones:<15} | "
              f"{self.anios:<10}")


# =========================================
# FUNCIONES DEL SISTEMA
# =========================================

lista_mecanicos = []


def agregar_mecanico():

    print("\n--- AGREGAR Clase Mecanico() ---")

    taller = input("Clase Mecanico() - Taller de trabajo: ")
    especialidad = input("Clase Mecanico() - Especialidad de motores: ")
    herramientas = input("Clase Mecanico() - Herramientas principales: ")
    certificaciones = input("Clase Mecanico() - Certificaciones: ")
    reparaciones = input("Clase Mecanico() - Reparaciones diarias: ")
    anios = input("Clase Mecanico() - Años de oficio: ")

    nuevo = Mecanico(
        taller,
        especialidad,
        herramientas,
        certificaciones,
        reparaciones,
        anios
    )

    lista_mecanicos.append(nuevo)

    print("\n✅ Clase Mecanico() agregada correctamente.\n")


def listar_mecanicos():

    print("\n================= LISTA Clase Mecanico() =================\n")

    if len(lista_mecanicos) == 0:

        print("No hay registros en Clase Mecanico().\n")

    else:

        print(f"{'No.':<5} | "
              f"{'TALLER':<20} | "
              f"{'ESPECIALIDAD':<20} | "
              f"{'HERRAMIENTAS':<20} | "
              f"{'CERTIFICACIONES':<20} | "
              f"{'REPARACIONES':<15} | "
              f"{'AÑOS':<10}")

        print("-" * 140)

        contador = 1

        for mecanico in lista_mecanicos:

            print(f"{contador:<5} | ", end="")
            mecanico.mostrar_datos()

            contador += 1

        print()


def actualizar_mecanico():

    listar_mecanicos()

    if len(lista_mecanicos) > 0:

        numero = int(input("Clase Mecanico() - Número a actualizar: "))

        if numero >= 1 and numero <= len(lista_mecanicos):

            mecanico = lista_mecanicos[numero - 1]

            print("\n--- ACTUALIZAR Clase Mecanico() ---")

            mecanico.taller = input("Clase Mecanico() - Nuevo taller: ")
            mecanico.especialidad = input("Clase Mecanico() - Nueva especialidad: ")
            mecanico.herramientas = input("Clase Mecanico() - Nuevas herramientas: ")
            mecanico.certificaciones = input("Clase Mecanico() - Nuevas certificaciones: ")
            mecanico.reparaciones = input("Clase Mecanico() - Nuevas reparaciones diarias: ")
            mecanico.anios = input("Clase Mecanico() - Nuevos años de oficio: ")

            print("\n✅ Clase Mecanico() actualizada correctamente.\n")

        else:

            print("\n❌ Número inválido.\n")


def borrar_mecanico():

    listar_mecanicos()

    if len(lista_mecanicos) > 0:

        numero = int(input("Clase Mecanico() - Número a borrar: "))

        if numero >= 1 and numero <= len(lista_mecanicos):

            lista_mecanicos.pop(numero - 1)

            print("\n🗑️ Clase Mecanico() eliminada correctamente.\n")

        else:

            print("\n❌ Número inválido.\n")


# =========================================
# MENÚ PRINCIPAL
# =========================================

opcion = 0

while opcion != 5:

    print("==============================================")
    print("      TALLER MECÁNICO INTELIGENTE")
    print("==============================================")
    print("1. Agregar Clase Mecanico()")
    print("2. Listar Clase Mecanico()")
    print("3. Actualizar Clase Mecanico()")
    print("4. Borrar Clase Mecanico()")
    print("5. Salir()")
    print("==============================================")

    opcion = int(input("Clase Mecanico() - Seleccione una opción: "))

    if opcion == 1:

        agregar_mecanico()

    elif opcion == 2:

        listar_mecanicos()

    elif opcion == 3:

        actualizar_mecanico()

    elif opcion == 4:

        borrar_mecanico()

    elif opcion == 5:

        print("\n👋 Programa Clase Mecanico() finalizado.")

    else:

        print("\n❌ Opción incorrecta.\n")
```
