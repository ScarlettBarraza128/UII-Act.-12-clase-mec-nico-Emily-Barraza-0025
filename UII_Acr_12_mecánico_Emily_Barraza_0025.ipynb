#### Problema Mecánico
* Taller de trabajo
* especialidad de motores
* herramientas principales
* certificaciones
* reparaciones diarias
* años de oficio.

```python
# =========================================
# PROGRAMA: TALLER MECÁNICO INTELIGENTE
# Todo el sistema dentro de class Mecanico()
# =========================================

class Mecanico():

    lista_mecanicos = []

    # =====================================
    # CONSTRUCTOR
    # =====================================

    def __init__(self, taller, especialidad, herramientas,
                 certificaciones, reparaciones, anios):

        self.taller = taller
        self.especialidad = especialidad
        self.herramientas = herramientas
        self.certificaciones = certificaciones
        self.reparaciones = reparaciones
        self.anios = anios

    # =====================================
    # MOSTRAR DATOS
    # =====================================

    def mostrar_datos(self):

        print(f"{self.taller:<20} | "
              f"{self.especialidad:<20} | "
              f"{self.herramientas:<20} | "
              f"{self.certificaciones:<20} | "
              f"{self.reparaciones:<15} | "
              f"{self.anios:<10}")

    # =====================================
    # AGREGAR MECÁNICO
    # =====================================

    def agregar_mecanico():

        print("\n--- AGREGAR Clase Mecanico() ---")

        taller = input("Clase Mecanico() - Taller de trabajo: ")
        especialidad = input("Clase Mecanico() - Especialidad: ")
        herramientas = input("Clase Mecanico() - Herramientas: ")
        certificaciones = input("Clase Mecanico() - Certificaciones: ")
        reparaciones = input("Clase Mecanico() - Reparaciones diarias: ")
        anios = input("Clase Mecanico() - Años de oficio: ")

        nuevo = Mecanico(
            taller,
            especialidad,
            herramientas,
            certificaciones,
            reparaciones,
            anios
        )

        Mecanico.lista_mecanicos.append(nuevo)

        print("\n✅ Clase Mecanico() agregada correctamente.\n")

    # =====================================
    # LISTAR MECÁNICOS
    # =====================================

    def listar_mecanicos():

        print("\n============= LISTA Clase Mecanico() =============\n")

        if len(Mecanico.lista_mecanicos) == 0:

            print("No hay registros.\n")

        else:

            print(f"{'No.':<5} | "
                  f"{'TALLER':<20} | "
                  f"{'ESPECIALIDAD':<20} | "
                  f"{'HERRAMIENTAS':<20} | "
                  f"{'CERTIFICACIONES':<20} | "
                  f"{'REPARACIONES':<15} | "
                  f"{'AÑOS':<10}")

            print("-" * 140)

            contador = 1

            for mecanico in Mecanico.lista_mecanicos:

                print(f"{contador:<5} | ", end="")
                mecanico.mostrar_datos()

                contador += 1

            print()

    # =====================================
    # ACTUALIZAR MECÁNICO
    # =====================================

    def actualizar_mecanico():

        Mecanico.listar_mecanicos()

        if len(Mecanico.lista_mecanicos) > 0:

            numero = int(input("Clase Mecanico() - Número a actualizar: "))

            if numero >= 1 and numero <= len(Mecanico.lista_mecanicos):

                mecanico = Mecanico.lista_mecanicos[numero - 1]

                print("\n--- ACTUALIZAR Clase Mecanico() ---")

                mecanico.taller = input("Nuevo taller: ")
                mecanico.especialidad = input("Nueva especialidad: ")
                mecanico.herramientas = input("Nuevas herramientas: ")
                mecanico.certificaciones = input("Nuevas certificaciones: ")
                mecanico.reparaciones = input("Nuevas reparaciones: ")
                mecanico.anios = input("Nuevos años de oficio: ")

                print("\n✅ Clase Mecanico() actualizada.\n")

            else:

                print("\n❌ Número inválido.\n")

    # =====================================
    # BORRAR MECÁNICO
    # =====================================

    def borrar_mecanico():

        Mecanico.listar_mecanicos()

        if len(Mecanico.lista_mecanicos) > 0:

            numero = int(input("Clase Mecanico() - Número a borrar: "))

            if numero >= 1 and numero <= len(Mecanico.lista_mecanicos):

                Mecanico.lista_mecanicos.pop(numero - 1)

                print("\n🗑️ Clase Mecanico() eliminada.\n")

            else:

                print("\n❌ Número inválido.\n")


# =========================================
# MENÚ PRINCIPAL
# =========================================

opcion = 0

while opcion != 5:

    print("==============================================")
    print("      TALLER MECÁNICO INTELIGENTE")
    print("==============================================")
    print("1. Agregar Clase Mecanico()")
    print("2. Listar Clase Mecanico()")
    print("3. Actualizar Clase Mecanico()")
    print("4. Borrar Clase Mecanico()")
    print("5. Salir()")
    print("==============================================")

    opcion = int(input("Clase Mecanico() - Seleccione una opción: "))

    if opcion == 1:

        Mecanico.agregar_mecanico()

    elif opcion == 2:

        Mecanico.listar_mecanicos()

    elif opcion == 3:

        Mecanico.actualizar_mecanico()

    elif opcion == 4:

        Mecanico.borrar_mecanico()

    elif opcion == 5:

        print("\n👋 Programa finalizado.")

    else:

        print("\n❌ Opción incorrecta.\n")
```
