#### Problema Mecánico
* Taller de trabajo
* especialidad de motores
* herramientas principales
* certificaciones
* reparaciones diarias
* años de oficio.

```python
# =========================================
# PROGRAMA: TALLER MECÁNICO INTELIGENTE
# Ejemplo de Programación Orientada a Objetos
# =========================================

class Mecanico:

    # Constructor
    def __init__(self, taller, especialidad, herramientas,
                 certificaciones, reparaciones, anios):

        self.taller = taller
        self.especialidad = especialidad
        self.herramientas = herramientas
        self.certificaciones = certificaciones
        self.reparaciones = reparaciones
        self.anios = anios

    # Mostrar información del mecánico
    def mostrar_datos(self):

        print(f"{self.taller:<20} | "
              f"{self.especialidad:<20} | "
              f"{self.herramientas:<20} | "
              f"{self.certificaciones:<20} | "
              f"{self.reparaciones:<15} | "
              f"{self.anios:<10}")


# =========================================
# FUNCIONES DEL SISTEMA
# =========================================

lista_mecanicos = []


def agregar_mecanico():

    print("\n--- AGREGAR MECÁNICO ---")

    taller = input("Taller de trabajo: ")
    especialidad = input("Especialidad de motores: ")
    herramientas = input("Herramientas principales: ")
    certificaciones = input("Certificaciones: ")
    reparaciones = input("Reparaciones diarias: ")
    anios = input("Años de oficio: ")

    nuevo = Mecanico(
        taller,
        especialidad,
        herramientas,
        certificaciones,
        reparaciones,
        anios
    )

    lista_mecanicos.append(nuevo)

    print("\n✅ Mecánico agregado correctamente.\n")


def listar_mecanicos():

    print("\n===================== LISTA DE MECÁNICOS =====================\n")

    if len(lista_mecanicos) == 0:

        print("No hay mecánicos registrados.\n")

    else:

        print(f"{'No.':<5} | "
              f"{'TALLER':<20} | "
              f"{'ESPECIALIDAD':<20} | "
              f"{'HERRAMIENTAS':<20} | "
              f"{'CERTIFICACIONES':<20} | "
              f"{'REPARACIONES':<15} | "
              f"{'AÑOS':<10}")

        print("-" * 140)

        contador = 1

        for mecanico in lista_mecanicos:

            print(f"{contador:<5} | ", end="")
            mecanico.mostrar_datos()

            contador += 1

        print()


def actualizar_mecanico():

    listar_mecanicos()

    if len(lista_mecanicos) > 0:

        numero = int(input("Número del mecánico a actualizar: "))

        if numero >= 1 and numero <= len(lista_mecanicos):

            mecanico = lista_mecanicos[numero - 1]

            print("\n--- NUEVOS DATOS ---")

            mecanico.taller = input("Nuevo taller: ")
            mecanico.especialidad = input("Nueva especialidad: ")
            mecanico.herramientas = input("Nuevas herramientas: ")
            mecanico.certificaciones = input("Nuevas certificaciones: ")
            mecanico.reparaciones = input("Nuevas reparaciones diarias: ")
            mecanico.anios = input("Nuevos años de oficio: ")

            print("\n✅ Mecánico actualizado correctamente.\n")

        else:

            print("\n❌ Número inválido.\n")


def borrar_mecanico():

    listar_mecanicos()

    if len(lista_mecanicos) > 0:

        numero = int(input("Número del mecánico a borrar: "))

        if numero >= 1 and numero <= len(lista_mecanicos):

            lista_mecanicos.pop(numero - 1)

            print("\n🗑️ Mecánico eliminado correctamente.\n")

        else:

            print("\n❌ Número inválido.\n")


# =========================================
# MENÚ PRINCIPAL
# =========================================

opcion = 0

while opcion != 5:

    print("====================================")
    print("   TALLER MECÁNICO INTELIGENTE")
    print("====================================")
    print("1. Agregar mecánico")
    print("2. Listar mecánicos")
    print("3. Actualizar mecánico")
    print("4. Borrar mecánico")
    print("5. Salir")
    print("====================================")

    opcion = int(input("Seleccione una opción: "))

    if opcion == 1:

        agregar_mecanico()

    elif opcion == 2:

        listar_mecanicos()

    elif opcion == 3:

        actualizar_mecanico()

    elif opcion == 4:

        borrar_mecanico()

    elif opcion == 5:

        print("\n👋 Programa finalizado.")

    else:

        print("\n❌ Opción incorrecta.\n")
```
