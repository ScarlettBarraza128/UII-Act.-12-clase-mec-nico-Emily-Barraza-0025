#### Problema Mecánico
* Taller de trabajo
* especialidad de motores
* herramientas principales
* certificaciones
* reparaciones diarias
* años de oficio.

```python
# =====================================================
# IA DE TALLER MECÁNICO
# Programación Orientada a Objetos
# Uso de:
# - Clase
# - Objetos
# - Métodos
# - Listas
# - if, elif, else
# - while True
# =====================================================

class Mecanico:

    # Lista general de mecánicos
    lista_mecanicos = []

    # =========================================
    # CONSTRUCTOR
    # =========================================
    def __init__(self, taller, especialidad, herramientas,
                 certificaciones, reparaciones, anios):

        self.taller = taller
        self.especialidad = especialidad
        self.herramientas = herramientas
        self.certificaciones = certificaciones
        self.reparaciones = reparaciones
        self.anios = anios

    # =========================================
    # AGREGAR MECÁNICO
    # =========================================
    def agregar_mecanico(self):

        Mecanico.lista_mecanicos.append(self)

        print("\n===================================")
        print("MECÁNICO AGREGADO CORRECTAMENTE")
        print("===================================")

    # =========================================
    # LISTAR MECÁNICOS
    # =========================================
    @classmethod
    def listar_mecanicos(cls):

        if len(cls.lista_mecanicos) == 0:
            print("\nNo hay mecánicos registrados.")
        else:

            print("\n================ LISTA DE MECÁNICOS ================\n")

            print("{:<5} {:<20} {:<22} {:<22} {:<18} {:<15} {:<10}".format(
                "No",
                "TALLER",
                "ESPECIALIDAD",
                "HERRAMIENTAS",
                "CERTIFICACIONES",
                "REPARACIONES",
                "AÑOS"
            ))

            print("-" * 120)

            contador = 1

            for mecanico in cls.lista_mecanicos:

                print("{:<5} {:<20} {:<22} {:<22} {:<18} {:<15} {:<10}".format(
                    contador,
                    mecanico.taller,
                    mecanico.especialidad,
                    mecanico.herramientas,
                    mecanico.certificaciones,
                    mecanico.reparaciones,
                    mecanico.anios
                ))

                contador += 1

    # =========================================
    # ACTUALIZAR MECÁNICO
    # =========================================
    @classmethod
    def actualizar_mecanico(cls):

        if len(cls.lista_mecanicos) == 0:
            print("\nNo hay mecánicos registrados.")
        else:

            cls.listar_mecanicos()

            numero = int(input("\nIngrese el número del mecánico a actualizar: "))

            if numero >= 1 and numero <= len(cls.lista_mecanicos):

                mecanico = cls.lista_mecanicos[numero - 1]

                print("\nIngrese los nuevos datos:\n")

                mecanico.taller = input("Nuevo taller: ")
                mecanico.especialidad = input("Nueva especialidad: ")
                mecanico.herramientas = input("Nuevas herramientas: ")
                mecanico.certificaciones = input("Nuevas certificaciones: ")
                mecanico.reparaciones = input("Reparaciones diarias: ")
                mecanico.anios = input("Años de oficio: ")

                print("\nMecánico actualizado correctamente.")

            else:
                print("\nNúmero inválido.")

    # =========================================
    # BORRAR MECÁNICO
    # =========================================
    @classmethod
    def borrar_mecanico(cls):

        if len(cls.lista_mecanicos) == 0:
            print("\nNo hay mecánicos registrados.")
        else:

            cls.listar_mecanicos()

            numero = int(input("\nIngrese el número del mecánico a borrar: "))

            if numero >= 1 and numero <= len(cls.lista_mecanicos):

                eliminado = cls.lista_mecanicos.pop(numero - 1)

                print("\nMecánico eliminado correctamente.")
                print("Especialidad eliminada:", eliminado.especialidad)

            else:
                print("\nNúmero inválido.")


# =====================================================
# MENÚ PRINCIPAL
# =====================================================

while True:

    print("\n====================================")
    print("      IA DE TALLER MECÁNICO")
    print("====================================")
    print("1. Agregar mecánico")
    print("2. Listar mecánicos")
    print("3. Actualizar mecánico")
    print("4. Borrar mecánico")
    print("5. Salir")
    print("====================================")

    opcion = input("Seleccione una opción: ")

    # =========================================
    # OPCIÓN 1
    # =========================================
    if opcion == "1":

        print("\n===== REGISTRO DE MECÁNICO =====\n")

        taller = input("Taller de trabajo: ")
        especialidad = input("Especialidad de motores: ")
        herramientas = input("Herramientas principales: ")
        certificaciones = input("Certificaciones: ")
        reparaciones = input("Reparaciones diarias: ")
        anios = input("Años de oficio: ")

        nuevo_mecanico = Mecanico(
            taller,
            especialidad,
            herramientas,
            certificaciones,
            reparaciones,
            anios
        )

        nuevo_mecanico.agregar_mecanico()

    # =========================================
    # OPCIÓN 2
    # =========================================
    elif opcion == "2":

        Mecanico.listar_mecanicos()

    # =========================================
    # OPCIÓN 3
    # =========================================
    elif opcion == "3":

        Mecanico.actualizar_mecanico()

    # =========================================
    # OPCIÓN 4
    # =========================================
    elif opcion == "4":

        Mecanico.borrar_mecanico()

    # =========================================
    # OPCIÓN 5
    # =========================================
    elif opcion == "5":

        print("\nGracias por usar la IA de Taller Mecánico.")
        break

    # =========================================
    # OPCIÓN INVÁLIDA
    # =========================================
    else:

        print("\nOpción inválida. Intente nuevamente.")
```
